```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('cleaned_titanic.csv')  # if saved
```

```python
df.head()
```

```python
# General overview of numerical features
print(df.describe())

# Overview of object-type (categorical) features
print(df.describe(include='object'))

# Check for unique values
print(df.nunique())
```

```python
df.hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.tight_layout()
plt.savefig('histogram_numerical_features.png')
plt.show()
```

```python
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplot of Age and Fare')
plt.savefig('boxplot_age_fare.png')
plt.show()
```

```python
categorical_cols = ['Pclass', 'Sex', 'Embarked']
for col in categorical_cols:
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.savefig(f'distribution_{col}.png')
    plt.show()
```
